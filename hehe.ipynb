```python
from bs4 import BeautifulSoup as bs

path = '.temp/Market.xml'

with open(path, 'r') as f:
    soup = bs(f, 'xml')

rows = soup.find_all('content')

for row in rows:
    cols = row.find('properties').find_all()

    for col in cols:
        print(col.name, col.text)
    
    
    

    


```
